# Предобработка

## Импорт библиотек

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

## Загрузка данных

```python
path2file = '../datasets/train.csv'
black_friday_df = pd.read_csv(path2file)
```

```python
path2file = '../datasets/test.csv'
test_black_friday_df = pd.read_csv(path2file)
```

## Предобработка

```python
black_friday_df = black_friday_df.drop(labels=['User_ID', 'Product_Category_2', 'Product_Category_3'], axis=1)
test_black_friday_df = test_black_friday_df.drop(labels=['User_ID', 'Product_Category_2', 'Product_Category_3'], axis=1)
```

```python
black_friday_df.Gender[black_friday_df.Gender == 'F'] = 1
black_friday_df.Gender[black_friday_df.Gender == 'M'] = 0

test_black_friday_df.Gender[test_black_friday_df.Gender == 'F'] = 1
test_black_friday_df.Gender[test_black_friday_df.Gender == 'M'] = 0
```

```python
black_friday_df.Age[black_friday_df.Age.isin(['0-17', '26-35', '46-50', '51-55', '36-45', '18-25'])] = 1
black_friday_df.Age[black_friday_df.Age == '55+'] = 0

test_black_friday_df.Age[test_black_friday_df.Age.isin(['0-17', '26-35', '46-50', '51-55', '36-45', '18-25'])] = 1
test_black_friday_df.Age[test_black_friday_df.Age == '55+'] = 0
```

```python
black_friday_df.City_Category[black_friday_df.City_Category.isin(['A', 'B'])] = 1
black_friday_df.City_Category[black_friday_df.City_Category == 'C'] = 0

test_black_friday_df.City_Category[test_black_friday_df.City_Category.isin(['A', 'B'])] = 1
test_black_friday_df.City_Category[test_black_friday_df.City_Category == 'C'] = 0
```

```python
black_friday_df.Stay_In_Current_City_Years[black_friday_df.Stay_In_Current_City_Years.isin(['0', '1', '2', '3'])] = 1
black_friday_df.Stay_In_Current_City_Years[black_friday_df.Stay_In_Current_City_Years == '4+'] = 0

test_black_friday_df.Stay_In_Current_City_Years[test_black_friday_df.Stay_In_Current_City_Years.isin(['0', '1', '2', '3'])] = 1
test_black_friday_df.Stay_In_Current_City_Years[test_black_friday_df.Stay_In_Current_City_Years == '4+'] = 0
```

```python
black_friday_df = pd.get_dummies(black_friday_df, prefix=['Occupation', 'Category'], columns=['Occupation', 'Product_Category_1'])

test_black_friday_df = pd.get_dummies(test_black_friday_df, 
                                      prefix=['Occupation', 'Category'], columns=['Occupation', 'Product_Category_1'])
```

```python
test_black_friday_df['Category_19'] = 0
test_black_friday_df['Category_20'] = 0
```
